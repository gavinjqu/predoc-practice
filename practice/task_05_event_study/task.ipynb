## 1. Setup & Data Generation

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
import statsmodels.formula.api as smf

plt.rcParams.update({
    'figure.figsize': (10, 6),
    'axes.spines.top': False,
    'axes.spines.right': False,
    'font.size': 12,
})
```

```python
# --- DATA GENERATION (do not modify) ---
np.random.seed(2024)

dates = pd.date_range('2019-01-01', '2022-12-31', freq='D')
n_days = len(dates)

# Events
events = {
    'Event_A': pd.Timestamp('2020-03-15'),  # Major recall
    'Event_B': pd.Timestamp('2021-08-22'),  # Moderate recall
    'Event_C': pd.Timestamp('2022-01-10'),  # Minor recall
}
event_effects = {
    'Event_A': {'num_reviews': 15, 'avg_rating': -0.3, 'num_complaints': 25, 'response_time_hrs': 4, 'negative_review_share': 0.12},
    'Event_B': {'num_reviews': 8, 'avg_rating': -0.15, 'num_complaints': 12, 'response_time_hrs': 2, 'negative_review_share': 0.06},
    'Event_C': {'num_reviews': 3, 'avg_rating': -0.05, 'num_complaints': 5, 'response_time_hrs': 0.5, 'negative_review_share': 0.02},
}

# Base series with trend, seasonality, and day-of-week effects
t = np.arange(n_days)
dow = pd.Series(dates).dt.dayofweek.values
dow_effect = np.where(dow >= 5, -5, 0)  # Weekend dip in reviews

num_reviews = 50 + 0.01 * t + 5 * np.sin(2 * np.pi * t / 365) + dow_effect + np.random.normal(0, 5, n_days)
avg_rating = 4.2 + 0.1 * np.sin(2 * np.pi * t / 365) + np.random.normal(0, 0.1, n_days)
num_complaints = 10 + 0.005 * t + np.random.poisson(3, n_days)
response_time = 12 + 2 * np.sin(2 * np.pi * t / 365) + np.random.exponential(2, n_days)
neg_share = 0.15 + 0.02 * np.sin(2 * np.pi * t / 365) + np.random.normal(0, 0.02, n_days)

# Apply event effects (persist for 30 days after event, then decay)
for event_name, event_date in events.items():
    effects = event_effects[event_name]
    for i, d in enumerate(dates):
        days_after = (d - event_date).days
        if 0 <= days_after <= 30:
            decay = 1 - (days_after / 60)  # Gradual decay
            num_reviews[i] += effects['num_reviews'] * decay
            avg_rating[i] += effects['avg_rating'] * decay
            num_complaints[i] += effects['num_complaints'] * decay
            response_time[i] += effects['response_time_hrs'] * decay
            neg_share[i] += effects['negative_review_share'] * decay

df = pd.DataFrame({
    'date': dates,
    'num_reviews': np.maximum(num_reviews, 0).round(0).astype(int),
    'avg_rating': avg_rating.clip(1, 5).round(2),
    'num_complaints': np.maximum(num_complaints, 0).round(0).astype(int),
    'response_time_hrs': np.maximum(response_time, 0.5).round(2),
    'negative_review_share': neg_share.clip(0, 1).round(4),
})

print(f"Data shape: {df.shape}")
print(f"Date range: {df['date'].min()} to {df['date'].max()}")
print(f"\nEvents:")
for name, date in events.items():
    print(f"  {name}: {date.date()}")
df.head()
```
