## 1. Setup & Data Generation

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

plt.rcParams.update({
    'figure.figsize': (10, 6),
    'axes.spines.top': False,
    'axes.spines.right': False,
    'font.size': 12,
})
```

```python
# --- DATA GENERATION (do not modify) ---
np.random.seed(123)

n_counties = 3200
regions = ['Northeast', 'South', 'Midwest', 'West']
state_names = [
    'Alabama', 'Alaska', 'Arizona', 'Arkansas', 'California', 'Colorado',
    'Connecticut', 'Delaware', 'Florida', 'Georgia', 'Hawaii', 'Idaho',
    'Illinois', 'Indiana', 'Iowa', 'Kansas', 'Kentucky', 'Louisiana',
    'Maine', 'Maryland', 'Massachusetts', 'Michigan', 'Minnesota',
    'Mississippi', 'Missouri', 'Montana', 'Nebraska', 'Nevada',
    'New Hampshire', 'New Jersey', 'New Mexico', 'New York',
    'North Carolina', 'North Dakota', 'Ohio', 'Oklahoma', 'Oregon',
    'Pennsylvania', 'Rhode Island', 'South Carolina', 'South Dakota',
    'Tennessee', 'Texas', 'Utah', 'Vermont', 'Virginia', 'Washington',
    'West Virginia', 'Wisconsin', 'Wyoming'
]
state_regions = {s: regions[i % 4] for i, s in enumerate(state_names)}

fips_codes = list(range(1001, 1001 + n_counties))

# Dataset 1: county_economics (wide, clean FIPS)
econ_data = pd.DataFrame({
    'county_fips': fips_codes,
    'median_income': np.random.normal(55000, 15000, n_counties).clip(20000, 150000).astype(int),
    'unemployment_rate': np.random.beta(2, 20, n_counties).round(4),
    'poverty_rate': np.random.beta(3, 15, n_counties).round(4),
    'population': np.random.lognormal(10, 1.2, n_counties).astype(int).clip(500, 5000000),
})
# Correlate poverty with income
econ_data['poverty_rate'] = (0.35 - econ_data['median_income'] / 500000 + np.random.normal(0, 0.03, n_counties)).clip(0.02, 0.45).round(4)

# Dataset 2: county_education (LONG format, some missing)
edu_rows = []
for fips in fips_codes:
    base = np.random.uniform(0.15, 0.45)
    for measure, val in [('pct_bachelors', base), 
                          ('pct_no_highschool', 0.5 - base + np.random.normal(0, 0.05)),
                          ('pct_graduate', base * 0.4 + np.random.normal(0, 0.03))]:
        # Randomly drop ~3% of values
        if np.random.random() > 0.03:
            edu_rows.append({'county_fips': fips, 'measure': measure, 'value': round(max(0, val), 4)})
edu_data = pd.DataFrame(edu_rows)

# Dataset 3: county_geography (some FIPS as strings with stripped zeros, extra counties)
geo_rows = []
for i, fips in enumerate(fips_codes):
    state = state_names[i % len(state_names)]
    # Randomly strip leading zeros for ~10% of FIPS (simulate formatting issue)
    fips_str = str(fips) if np.random.random() > 0.1 else str(int(str(fips).lstrip('0')))
    geo_rows.append({
        'county_fips': fips_str,
        'county_name': f'County_{i+1}',
        'state_name': state,
        'region': state_regions[state],
    })
# Add 50 extra counties not in other datasets
for j in range(50):
    geo_rows.append({
        'county_fips': str(99001 + j),
        'county_name': f'Extra_County_{j+1}',
        'state_name': state_names[j % len(state_names)],
        'region': state_regions[state_names[j % len(state_names)]],
    })
geo_data = pd.DataFrame(geo_rows)

print("county_economics:", econ_data.shape)
print("county_education:", edu_data.shape)
print("county_geography:", geo_data.shape)
```
