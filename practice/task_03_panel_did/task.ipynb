## 1. Setup & Data Generation

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
import statsmodels.formula.api as smf

plt.rcParams.update({
    'figure.figsize': (10, 6),
    'axes.spines.top': False,
    'axes.spines.right': False,
    'font.size': 12,
})
```

```python
# --- DATA GENERATION (do not modify) ---
np.random.seed(42)

n_states = 50
quarters = pd.date_range('2017-01-01', '2022-12-31', freq='QS')
n_quarters = len(quarters)
treated_states = list(range(1, 11))  # States 1-10 are treated

rows = []
for s in range(1, n_states + 1):
    # State-level baseline
    base_emp = np.random.uniform(85, 105)
    base_pop = np.random.uniform(500, 5000)
    base_wage = np.random.uniform(10, 18)
    rest_share = np.random.uniform(0.05, 0.15)
    is_treated = 1 if s in treated_states else 0
    
    for i, q in enumerate(quarters):
        post = 1 if q >= pd.Timestamp('2020-01-01') else 0
        # Time trend + seasonal
        trend = 0.3 * i
        seasonal = 2 * np.sin(2 * np.pi * i / 4)
        # Treatment effect: -3.5 units for treated states post-treatment
        effect = -3.5 * is_treated * post
        # Add noise
        noise = np.random.normal(0, 2)
        
        employment = base_emp + trend + seasonal + effect + noise
        population = base_pop + np.random.normal(0, 50)
        avg_wage = base_wage + 0.1 * i + np.random.normal(0, 0.5)
        r_share = rest_share + np.random.normal(0, 0.005)
        
        rows.append({
            'state_id': s,
            'quarter': q,
            'treated': is_treated,
            'post': post,
            'employment': round(employment, 1),
            'population': round(population, 0),
            'avg_wage': round(avg_wage, 2),
            'restaurant_share': round(r_share, 4),
        })

df = pd.DataFrame(rows)
print(f"Panel shape: {df.shape}")
print(f"States: {df['state_id'].nunique()}, Quarters: {df['quarter'].nunique()}")
print(f"Treated: {df[df['treated']==1]['state_id'].nunique()}, Control: {df[df['treated']==0]['state_id'].nunique()}")
df.head()
```
