### DAY 7 — Groupby, Aggregation, Multi-Agg, Custom Functions
**Core Drill A — Basic Aggregations**
- Compute in one line each:

1. Mean BMI by gender
2. Median triglycerides by ethnicity
3. Count of diagnosed diabetics by age group (use cut/bins)

````python
mean_bmi_by_gender = df.groupby('gender')['bmi'].mean()

median_triglyceride_by_eth = df.groupby('ethnicity')['triglycerides'].median()

bins = [0, 13, 18, 65, 100]
labels = ['Child', 'Teenager', 'Adult', 'Senior']
df['age_group'] = pd.cut(df['age'], bins=bins, labels=labels, right=False)

count_diagnosed_by_age = df.groupby('age_group')['diagnosed_diabetes'].count()

print(f"Mean BMI by gender: {mean_bmi_by_gender}")
print(f"Median tryglyceride by ethnicity: {median_triglyceride_by_eth}")
print(f"count of diagnosed diabetics by age: {count_diagnosed_by_age}")
````

**Core Drill B — Multi-Aggregation**
- What is the resulting column index structure?
- How do you flatten the MultiIndex?

````python
multi_agg_df = df.groupby("ethnicity").agg({
    "bmi": ["mean", "std"],
    "hba1c": ["mean", "max"],
    "cholesterol_total": "median"
})

cols = multi_agg_df.columns
multi_agg_df.columns = ['_'.join(col).strip() for col in cols]
print(multi_agg_df)
````

**Core Drill C — Custom Lambda Functions**
1. Coefficient of variation of BMI by ethnicity:
``lambda x: x.std() / x.mean()``
2. Percentage of population with diagnosed diabetes within each education level
3. Average metabolic “cluster score” via:
``(bmi + waist_to_hip_ratio + triglycerides)``

````python
bmi_cv = df.groupby('ethnicity')['bmi'].agg(
    cv_bmi=lambda x: x.std() / x.mean()     # x is series of all bmi for each ethnic group
)

diabetes_pct = df.groupby('education_level')['diagnosed_diabetes'].agg(
    pct_diabetes=lambda x: (x.mean() * 100).round(2)
)

# create cluster score and aggregate for mean
df['meta_cluster_score'] = (
    df['bmi'] + df['waist_to_hip_ratio'] + df['triglycerides']
)
avg_meta_score = df.groupby('ethnicity')['meta_cluster_score'].mean()   
avg_metabolic = df.groupby('ethnicity')['meta_cluster_score'].agg(
    mean_meta_score=lambda x: x.mean()
)

print(bmi_cv)
print(diabetes_pct)
print(avg_meta_score)
print(avg_metabolic)
````
